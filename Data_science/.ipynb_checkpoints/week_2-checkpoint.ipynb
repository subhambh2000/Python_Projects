```python
test = pd.read_csv('olympics.csv')
test
```

```python
test = pd.read_csv('olympics.csv', index_col=0, skiprows=1)
test.head()
```

```python
test.columns
```

```python
for col in test.columns:
    if col[:2] == '01':
        test.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2] == '02':
        test.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2] == '03':
        test.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1] == '№':
        test.rename(columns={col:'#'+col[1:]}, inplace=True)
```

```python
test.head()
```

```python
cen = pd.read_csv('census.csv')
cen.head()
```

```python
cen['SUMLEV'].unique()
```

```python
cen = cen[cen['SUMLEV'] == 50]
cen.head()
```

```python
columns_to_keep = ['STNAME',
                                  'CTYNAME',
                                  'BIRTHS2010',
                                  'BIRTHS2011',
                                  'BIRTHS2012',
                                  'BIRTHS2013',
                                  'BIRTHS2014',
                                  'BIRTHS2015',
                                  'POPESTIMATE2010',
                                  'POPESTIMATE2011',
                                  'POPESTIMATE2012',
                                  'POPESTIMATE2013',
                                  'POPESTIMATE2014',
                                  'POPESTIMATE2015']
cen = cen[columns_to_keep]
cen.head()
```

```python
cen = cen.set_index(['STNAME', 'CTYNAME'])
cen.head()
```

```python
cen.loc['Michigan', 'Washtenaw County']
```

```python
# cen.loc[[('Michigan', 'Washtenaw County'), ('Michigan', 'Wayne County')]] this method is deprecated  the alternative is to use reindex as used below
cen.reindex([('Michigan', 'Washtenaw County'), ('Michigan', 'Wayne County')])
```
